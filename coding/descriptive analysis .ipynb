```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Sample dataset: Monthly Salaries (in $1000s) of 20 employees
data = pd.Series([30, 35, 40, 40, 45, 50, 55, 55, 55, 60, 
                  65, 70, 75, 80, 85, 90, 95, 100, 110, 120])

# 1ï¸âƒ£ Mean (Average Salary)
mean_value = np.mean(data)

# 2ï¸âƒ£ Median (Middle Salary)
median_value = np.median(data)

# 3ï¸âƒ£ Mode (Most Common Salary) - Updated for SciPy 1.11+
mode_value = stats.mode(data, keepdims=True).mode[0]

# 4ï¸âƒ£ Variance (Spread of Salaries)
variance_value = np.var(data, ddof=1)

# 5ï¸âƒ£ Standard Deviation (How much salaries deviate from the mean)
std_dev_value = np.std(data, ddof=1)

# 6ï¸âƒ£ Skewness (Symmetry of Salary Distribution)
skewness_value = stats.skew(data)

# 7ï¸âƒ£ Kurtosis (Peakedness of Salary Distribution)
kurtosis_value = stats.kurtosis(data)

# 8ï¸âƒ£ Percentiles (25th, 50th, 75th)
percentiles = np.percentile(data, [25, 50, 75])

# 9ï¸âƒ£ Interquartile Range (IQR)
IQR_value = percentiles[2] - percentiles[0]

# Display all results
print(f"ðŸ“Œ Mean Salary: ${mean_value:.2f}K")
print(f"ðŸ“Œ Median Salary: ${median_value}K")
print(f"ðŸ“Œ Mode Salary: ${mode_value}K")
print(f"ðŸ“Œ Variance: {variance_value:.2f}")
print(f"ðŸ“Œ Standard Deviation: {std_dev_value:.2f}")
print(f"ðŸ“Œ Skewness: {skewness_value:.2f} (Right-Skewed if > 0, Left-Skewed if < 0)")
print(f"ðŸ“Œ Kurtosis: {kurtosis_value:.2f} (High Kurtosis = More Outliers)")
print(f"ðŸ“Œ 25th Percentile (Q1): ${percentiles[0]}K")
print(f"ðŸ“Œ 50th Percentile (Median): ${percentiles[1]}K")
print(f"ðŸ“Œ 75th Percentile (Q3): ${percentiles[2]}K")
print(f"ðŸ“Œ Interquartile Range (IQR): ${IQR_value}K")

# ðŸ“Š **Visualizing Data**
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# Histogram for Salary Distribution
sns.histplot(data, bins=8, kde=True, color="skyblue", edgecolor="black", ax=axes[0])
axes[0].axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value}')
axes[0].axvline(median_value, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_value}')
axes[0].legend()
axes[0].set_title("Salary Distribution")

# Boxplot for Outliers
sns.boxplot(data, color="orange", notch=True, ax=axes[1])
axes[1].set_title("Boxplot: Detecting Outliers")

plt.tight_layout()
plt.show()
```
