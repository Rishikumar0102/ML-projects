```python
import google.generativeai as genai
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langgraph.graph import StateGraph

# ‚úÖ Step 1: Configure Gemini API
api_key = "AIzaSyBNX8KAXoSP2x3JejdKEX_Dit48YVJR1xg"  # Replace with your actual key
genai.configure(api_key=api_key)

# ‚úÖ Step 2: Initialize LangChain's Gemini model
gemini = ChatGoogleGenerativeAI(model="gemini-2.0-flash", google_api_key=api_key)

# ‚úÖ Step 3: Define a function to summarize text using Gemini
def summarize_text(text):
    prompt = f"Summarize the following text in 3 lines:\n\n{text}"
    response = gemini.invoke(prompt)
    return response.content.strip()

# ‚úÖ Step 4: Function to perform multi-step summarization
def multi_step_summarizer(article):
    # Step 4.1: Split the article into sections
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
    sections = text_splitter.split_text(article)

    # Step 4.2: Summarize each section
    summaries = [summarize_text(section) for section in sections]

    # Step 4.3: Combine all section summaries into a final summary
    final_summary = summarize_text(" ".join(summaries))

    return final_summary

# ‚úÖ Step 5: Define a Sample Long Article
long_article = """
Artificial Intelligence (AI) is a field of computer science that focuses on creating systems that can perform 
tasks that typically require human intelligence. It includes machine learning, natural language processing, and 
robotics. Machine learning allows computers to learn from data and make decisions without explicit programming. 
Natural language processing (NLP) enables machines to understand human language. Robotics combines AI with 
mechanical engineering to create intelligent robots. AI has applications in healthcare, finance, and many other 
fields. However, it also raises ethical concerns regarding privacy and job displacement.
"""

# ‚úÖ Step 6: Run the Multi-Step Summarizer
final_summary = multi_step_summarizer(long_article)

# ‚úÖ Step 7: Print the Final Summary
print("üîπ Final Multi-Step Summary:")
print(final_summary)
```
