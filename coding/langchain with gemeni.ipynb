```python
from langchain_google_genai import ChatGoogleGenerativeAI  
from langchain_core.messages import HumanMessage  
import json  

# Initialize the Gemini model
llm = ChatGoogleGenerativeAI(model="gemini-pro", google_api_key="AIzaSyBaY0u_-2SMmUNg_SDDBqYJl11PFPvid9Y")

for i in range(10):  
    question_prompt = """  
    You are an HR bot for Mindful AI.  
    Generate ONE multiple-choice question for a Machine Learning role.  
    The response must be in JSON format like this:  
    {  
      "question": "What is the purpose of a loss function in Machine Learning?",  
      "options": {  
        "a": "To increase model accuracy",  
        "b": "To measure the error",  
        "c": "To select the best optimizer"  
      },  
      "answer": "b"  
    }  
    """  

    # Generate response
    response = llm.invoke([HumanMessage(content=question_prompt)])

    # Parse JSON output
    try:
        question_data = json.loads(response.content)
        question_text = question_data["question"]
        options = question_data["options"]
        
        print(f"Q{i+1}: {question_text}")
        for key, value in options.items():
            print(f"  {key}) {value}")
    except json.JSONDecodeError:
        print("Error: Could not parse the response as JSON.")

```
