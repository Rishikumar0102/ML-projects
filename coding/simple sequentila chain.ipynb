```python
from langchain_google_genai import GoogleGenerativeAI

from langchain.chains import SimpleSequentialChain
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Initialize Gemini 2.0 Flash
GOOGLE_API_KEY = "AIzaSyBJm0Zhit7_tpitUJ58hDu758RVHiR3BDo"
llm = GoogleGenerativeAI(model="gemini-2.0-flash", api_key=GOOGLE_API_KEY)

# Chain 1: Generate a fun fact about a given topic
fact_prompt = PromptTemplate.from_template("Tell me who will find one piece {topic}.")
fact_chain = LLMChain(llm=llm, prompt=fact_prompt)

# Chain 2: Generate a short joke based on the fact
joke_prompt = PromptTemplate.from_template("Create a funny joke based on this fact: {text}")
joke_chain = LLMChain(llm=llm, prompt=joke_prompt)

# Combine using SimpleSequentialChain
simple_chain = SimpleSequentialChain(
    chains=[fact_chain, joke_chain], verbose=True
)

# Run the chain with input
result = simple_chain.run("one piece ")

print("\nFinal Joke:\n", result)
```
