```python
from langchain_google_genai import GoogleGenerativeAI
from langchain.chains import SequentialChain
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# 🔑 Set up Gemini 2.0 Flash LLM
GOOGLE_API_KEY = "AIzaSyBJm0Zhit7_tpitUJ58hDu758RVHiR3BDo"
llm = GoogleGenerativeAI(model="gemini-2.0-flash", google_api_key=GOOGLE_API_KEY)

# ✅ Chain 1: Generate a creative story title
title_prompt = PromptTemplate.from_template("Suggest a creative short story title about {genre}.")
title_chain = LLMChain(llm=llm, prompt=title_prompt, output_key="title")

# ✅ Chain 2: Generate a short story based on the title
story_prompt = PromptTemplate.from_template("Write a short story with the title: {title}.")
story_chain = LLMChain(llm=llm, prompt=story_prompt, output_key="story")

# ✅ Chain 3: Generate a moral based on the story
moral_prompt = PromptTemplate.from_template("What is the moral of this story? {story}")
moral_chain = LLMChain(llm=llm, prompt=moral_prompt, output_key="moral")

# 🔗 Combine all chains using SequentialChain
sequential_chain = SequentialChain(
    chains=[title_chain, story_chain, moral_chain], 
    input_variables=["genre"], 
    output_variables=["title", "story", "moral"], 
    verbose=True
)

# 🎯 Run the chain
result = sequential_chain({"genre": "science fiction"})

# 🖨️ Print results
print("\nGenerated Title:", result["title"])
print("\nGenerated+Story:", result["story"])
print("\nStory Moral:", result["moral"])
```
